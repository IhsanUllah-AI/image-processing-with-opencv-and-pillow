Histogram Equalization is a computer image processing technique used to improve contrast in images
load image using python libraries

```python
import numpy as np
#matplotlib
import matplotlib.pyplot as plt
#opencv
import cv2
#load image and convert to gray scale
img=cv2.imread('b.jpg',cv2.IMREAD_GRAYSCALE)
cv2.namedWindow('original image',cv2.WINDOW_NORMAL) #by default image appear zoom in 
#imshow take 2 argument first window name on whic image is opened 2nd is image array
cv2.imshow('original image',img) #imshwo must take first argument is window name if you dont specifed img will consider the window
cv2.waitKey(0)
cv2.destroyAllWindows()
print("image load and been displayed ")
```

calculate histogram of original image 
calchist take 5 parameter first is img array ,channel,mask,max numb of intensity,range of intensity
apply histogram function

```python
hist=cv2.calcHist([img],[0],None,[256],[0,256])
#hist is list of size 256 from 0 to 255 count frequency
#plot it 
plt.title("original image intensity")
plt.xlabel("intensity")
plt.ylabel("Frequency")
plt.xlim([0,256])
plt.plot(hist,color='blue')
plt.show()
```

apply histogram equalization to enhance image brightness  whuch is 4 step cumsum,mult with max intensity,round off,replace with new intensity

```python
img_eq=cv2.equalizeHist(img)
histog=cv2.calcHist(img_eq,[0],None,[256],[0,256])
#plot it 
plt.title("normalize image intensity")
plt.xlabel("intensity")
plt.ylabel("Frequency")
plt.xlim([0,256])
plt.plot(histog,color='blue')
plt.show()
```

imshwo must take first argument is window name if you dont specifed img will consider the window

```python
cv2.namedWindow('equalized image',cv2.WINDOW_NORMAL) 
cv2.imshow('equalized image',img_eq) 
cv2.waitKey(0)
cv2.destroyAllWindows()
print("new image has been disalyed")
```
